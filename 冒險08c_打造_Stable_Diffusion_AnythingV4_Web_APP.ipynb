### 2. `pipeline` 一條龍服務

```python
import torch

from diffusers import StableDiffusionPipeline
```

```python
pipe = StableDiffusionPipeline.from_pretrained("andite/anything-v4.0", 
                                               torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

### 3. 用 `gradio` 打造你的 Web App!

```python
import gradio as gr
```

```python
def text2image(prompt, negative_prompt):
    image = pipe(prompt, negative_prompt=negative_prompt).images[0]
    return image
```

```python
inp1 = gr.Textbox(label="輸入你的咒語 (prompt)")
inp2 = gr.Textbox(label="負向的咒語 (negative prompt)")
outputs = gr.Image(label="生成圖")
```

可以自己設範例!

```python
prompt1 = "a cute girl teaching in front of a chalkboard"
prompt2 = "a cute (owl), flat 2d, simple vector art, minimalist"
prompt3 = "collection of icon designs for (animals), flat 2d, simple vector art"
prompt4 = "a cute obese round (tiger), minimalist, pixar style"
negative_prompt = ""
```

```python
examples = [[prompt1, negative_prompt],
            [prompt2, negative_prompt],
            [prompt3, negative_prompt],
            [prompt4, negative_prompt]]
```

```python
iface = gr.Interface(text2image,
                     inputs=[inp1, inp2],
                     outputs=outputs,
                     title="Stable Diffusion AnitmationV4 文字生圖 AI",
                     description="請輸入一段英文咒語 (prompt), 我會根據你的咒語畫出一張圖!",
                     examples=examples)
```

```python
iface.launch(share=True)
```
